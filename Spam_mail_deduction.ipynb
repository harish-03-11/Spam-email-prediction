```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/spam.csv")
```

```python
df.sample(3)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df['target'] = df['Category'].apply(lambda x : 1 if x=='spam' else 0 )
```

```python
df.sample(5)
```

```python
df.drop("Category", axis = 1)
```

```python
X = df['Message']
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25, random_state = 5)
```

```python
v = CountVectorizer()
```

```python
x_train_values = v.fit_transform(X_train.values)
```

```python
print(x_train_values.toarray()[:2])
```

```python
model = MultinomialNB()
```

```python
x_train_nb = model.fit(x_train_values, y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'there you go click on this link for bonus cash rewards'
]
```

```python
email_values = v.transform(emails)
```

```python
model.predict(email_values)
```
